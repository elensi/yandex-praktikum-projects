# [Описание проекта](#Содержание) <a class="ancor" id="Описание_проекта"></a> 

В данном проекте мы исследуем отток клиентов в сети фитнес-центров. 

Одна из распространенных проблем фитнес-клубов - отток клиентов. Не всегда легко понять, что клиент больше не с вами. Можно записать в отток тех, кто попросил закрыть договор или удалил аккаунт. Однако клиенты не всегда уходят демонстративно: чаще перестают пользоваться сервисом тихо. Индикаторы оттока зависят от специфики отрасли.  

Для фитнес-центра можно считать, что клиент попал в отток, если за последний месяц ни разу не посетил спортзал.  

Для борьбы с оттоком, отдел по работе с клиентами перевел в электронный вид большинство анкет пользователей. 

Наша задача - провести анализ и подготовить план действий по удержанию клиентов. А именно:

- спрогнозировать вероятность оттока (на уровне следующего месяца) для каждого клиента;
- сформировать типичные портреты пользователей: выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;
- проанализировать основные признаки, наиболее сильно влияющие на отток;
- сформулировать основные выводы и разработать рекомендации по повышению качества работы с клиентами:
1) выделить целевые группы клиентов;
2) предложить меры по снижению оттока;
3) определить другие особенности взаимодействия с клиентами.

# [Модель прогнозирования оттока пользователей](#Содержание) <a class="ancor" id="Шаг_3"></a>

Для прогнозирования оттока пользователей построим модель бинарной классификации, где целевой признак - акт оттока пользователя в следующем месяце.  

Модель обучим 2 способами:
- логистической регрессией
- случайным лесом.

Для оценки сравним метрики:
- Доля правильных ответов (англ. accuracy),
- Точность (англ. precision),
- Полнота (англ. recall)

```python
# список моделей, которые мы будем использовать
models = [LogisticRegression(), RandomForestClassifier()]
```

```python
# разделим наши данные на признаки (матрица X) и целевую переменную (y)
X = gym_churn.drop('churn', axis = 1)
y = gym_churn['churn']
```

```python
# разделим модель на обучающую и валидационную выборку
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# стандартизируем данные 
scaler = StandardScaler()
X_train_st = scaler.fit_transform(X_train) #обучаем scaler и одновременно трансформируем матрицу для обучающей выборки
X_test_st = scaler.transform(X_test)# стандартизируйте тестовую выборку методом transform scaler
```

```python
# определим функцию, которая будет выводить наши метрики
def print_all_metrics(y_true, y_pred, y_proba, title = 'Метрики классификации'):
    print(title)
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred)))
```

### [Логистическая регрессия](#Содержание) <a class="ancor" id="Шаг_31"></a>

```python
# зададим алгоритм для модели логистической регрессии
lr_model = LogisticRegression()
# обучим модель
lr_model.fit(X_train_st, y_train)
# воспользуемся уже обученной моделью, чтобы сделать прогнозы
lr_predictions = lr_model.predict(X_test_st)
lr_probabilities = lr_model.predict_proba(X_test_st)[:,1]
# выведем все метрики
print_all_metrics(y_test, lr_predictions, lr_probabilities , title='Метрики для модели логистической регрессии:')
```

### <font color=Teal>[Случайный лес](#Содержание) <a class="ancor" id="Шаг_32"></a>

```python
# зададим алгоритм для новой модели на основе алгоритма случайного леса
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 0) # Ваш код здесь
# обучим модель случайного леса
rf_model.fit(X_train_st, y_train)
# воспользуемся уже обученной моделью, чтобы сделать прогнозы
rf_predictions = rf_model.predict(X_test_st)# Ваш код здесь
rf_probabilities = rf_model.predict_proba(X_test_st)[:,1]# Ваш код здесь
# выведем все метрики
print_all_metrics(y_test, rf_predictions, rf_probabilities, title = 'Метрики для модели случайного леса:')
```

Обе модели показали себя хорошо. Доля правильных ответов в обоих случаях 0,92. Немного лучше себя показала модель логистической регрессии в метриках  точности и полноты. Метрика precision(точность) говорит, какая доля прогнозов относительно "1" класса верна. То есть смотрим долю правильных ответов только среди целевого класса, которая свидетельствует об оттоке клиентов. Поэтому правильнее будет выбрать первую модель.

# [Кластеризация пользователей](#Содержание) <a class="ancor" id="Шаг_4"></a>

### Стандартизация данных

```python
scaler = StandardScaler() # создаём объект класса scaler (нормализатор)
x_sc = scaler.fit_transform(gym_churn) # обучаем и преобразуем набор данных
```

### [Агломеративная иерархическая кластеризация ](#Содержание) <a class="ancor" id="Шаг_41"></a>

```python
linked = linkage(x_sc, method = 'ward')
```

```python
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.title('Hierarchial clustering for GYM')
plt.show()
```

На основании построенной дендрограммы можно предположить, что  оптимальное число кластеров 5 — пять разных цветов на графике.

### [K-Means ](#Содержание) <a class="ancor" id="Шаг_42"></a>

```python
km = KMeans(n_clusters = 5) # задаём число кластеров, равное 5
# прогнозируем кластеры для наблюдений (алгоритм присваивает им номера от 0 до 4)
labels = km.fit_predict(x_sc)
```

```python
# сохраняем метки кластера в поле нашего датасета
gym_churn['cluster_km'] = labels
```

```python
# выводим статистику по средним значениям наших признаков по кластеру
gym_churn.groupby('cluster_km').mean()
```

Анализируя средние значения признаков в разных кластерах можно заметить, что в 0 кластере клиенты с длинным сроком действия карты, также большая доля карт партнеров и приглашений промо.
1 кластер - клиенты, которые чаще всех ходят в зал  
3 кластер - ушедшие клиенты, значение оттока 0,93, это значит что почти все клиент в данном кластере перестали посещать спортзал.  
2 кластер - клиенты, живущие или работающие далеко от фитнес-клуба.  
4 кластер - не указан номер телефона

### [Распределение признаков по кластерам](#Содержание) <a class="ancor" id="Шаг_43"></a>

```python

for feature in gym_churn.columns:
    print('Распределение',feature, 'клиентов кластеров')
    gym_churn.query('cluster_km == "4"')[feature].hist(alpha=0.5)
    gym_churn.query('cluster_km == "3"')[feature].hist(alpha=0.5)
    gym_churn.query('cluster_km == "2"')[feature].hist(alpha=0.5)
    gym_churn.query('cluster_km == "0"')[feature].hist(alpha=0.5)
    gym_churn.query('cluster_km == "1"')[feature].hist(alpha=0.5)
 
    
    plt.xlabel("Значение признака")
    plt.ylabel("Частота")
    plt.legend(labels=gym_churn['cluster_km'].unique())
    plt.show()
```

#### Основные отличительные характеристики кластеров

- 0 кластер - клиенты партнеров, долгий срок действия карты.
- 1 кластер - клиенты с высоким lifetime  и возрастом клиентов
- 2 кластер -клиенты, живущие или работающие далеко от фитнес-клуба.  
- 3 кластер - ушедшие клиенты, значение оттока 0,99, это значит что почти все клиент в данном кластере перестали посещать спортзал.  
- 4 кластер - клиенты без номера телефона.  

Отдельно стоит расммотреть характеристики кластера со значение оттока 0,93. Это в основном клиенты, живущие рядом с фитнес-клубом, средний возраст 26 лет, они недолго пользуются услугами клуба и бросают посещение. Не тратят деньги в клубе. Возможно в первый месяц клуб их не смог заинтересовать.

### [Доля оттока](#Содержание) <a class="ancor" id="Шаг_44"></a>

```python
#всего в группе
churn_rate_all = gym_churn.groupby(['cluster_km'])['churn'].count()
```

```python
#ушли
churn_rate=gym_churn.groupby(['cluster_km','churn'])['age'].count().reset_index().query('churn == "1"')
```

```python
#вычислим долю ушедших в каждом кластере
churn_rate= churn_rate.merge(churn_rate_all, on = 'cluster_km')
churn_rate.columns = ['cluster_km', '1', 'churn', 'all']
churn_rate = churn_rate[['cluster_km','churn', 'all']]
churn_rate['churn_rate'] = churn_rate['churn'] / churn_rate['all']
churn_rate
```

Как и было уже озвучено выше, кластер под номером 3  - кластер ушедших клиентов. В остальных кластеров доля оттока минимальна. Стоит обратить внимание на 2 кластер - в нем доля оттока 43%. Отличительной особенностью данного кластера является то, что клиенты живут или работают далеко от фитнес-клуба. Видимо данный фактор влияет на решение клиента продолжать посещение клуба или нет.
